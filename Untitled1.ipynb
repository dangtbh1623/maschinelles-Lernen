 **Benötigte Biliotheken importieren

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import numpy as np
```

**Daten erstellen

```python
data=np.array([[1,1],[2,2],[3,3],[1,5],[5,4],])
labelList = ['A','B','C','D','E']
```

**Verwendet Clustering-Ansatz Single-Linkage mit der Minkowski Metrik p=1

```python
linked = linkage(data, 'single', metric='cityblock')
```

**Zeichne Dendrogramm

```python
plt.figure(figsize=(10, 7))
plt.title('Iris Hierarchical Clustering Dendrogram')
plt.xlabel('Species')
plt.ylabel('distance')
dendrogram(linked,
          labels=labelList,
           distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

**Zeichne der Bereich, in dem man 3 Cluster erhalten kann

```python
plt.figure(figsize=(10, 7))
dendrogram(linked,
          labels=labelList,
           distance_sort='descending',
            show_leaf_counts=True)
plt.axhline(y=2.5, c='red')
plt.show()
```
