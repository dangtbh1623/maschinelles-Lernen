**Laden Iris-Datensatz

```python
from sklearn import datasets
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
iris = datasets.load_iris()
X=iris.data[1::3,:]
Y=iris.target[1::3]
```

**Die Blumen zu Cluster einordnen. Bei DEFAULT werden die Methode "Single" und Distanzma√ü "euclidean" eingestellt.
Mit der "method=" und "metric=" kann man verschiedenen Einstellungen verwenden.

```python
linked = linkage(X, method='ward',metric='euclidean')
```

**Plotten

```python
plt.figure(figsize=(20, 10))
plt.title('Iris Hierarchical Clustering Dendrogram')
plt.xlabel('Species')
plt.ylabel('Distance')
dendrogram(linked, labels=Y, leaf_rotation=0., leaf_font_size=16., distance_sort='descending' )
plt.show()
```
