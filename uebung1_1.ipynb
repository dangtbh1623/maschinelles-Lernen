# Import Library and Data

```python
from scipy.io import arff
import pandas as pd
import numpy as np
data = arff.loadarff('U:/Dortmund/Master/ML/breast-w.arff')
df = pd.DataFrame(data[0])
```

Konvertiert Class-Objekte zur Str UTF-8

```python
df['Class'] = df['Class'].str.decode('utf-8')
```

## Was ist die Klassenverteilung der beiden Krebsarten (gutartig / bösartig) in den Daten?

```python
df['Class'].value_counts()
print(df['Class'].value_counts())
classes, counts = np.unique(df['Class'], return_counts=True)
print("\nKlassen:", classes, "Beobachtungen:", counts)
```
